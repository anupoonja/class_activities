```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
# When can you use on in pd.merge?
# the column names between left and right dataframes must be exactly same
merged_df = pd.merge(bitcoin_df, dash_df, how="inner", on=["Date","Date"], suffixes=('_b','_d'))
merged_df.head()
```

```python
# Collecting best open for Bitcoin and Dash

best_opending_ = merged_df.sort_values(["Open_b"], ascending=False)
#best_opending.head()

# Collecting best close for Bitcoin and Dash
best_closing = merged_df.sort_values(["Close_d"], ascending=False)
best_closing.head()

dash_close = merged_df["Open_d"].max()
dash_close

# Collecting the total volume for Bitcoin and Dash
dash_volume = merged_df["Volume_d"].sum()
dash_volume
```
